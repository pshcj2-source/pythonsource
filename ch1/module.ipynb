```python
import random
#random : 난수(규칙이 없는 임의의 수를 발생시키는 모듈)

print(random.random())         #random() : 0.0 ~ 1.0 사이의 실수 중에서 난수 (0은 포함하지만 1은 미포함)
# 0.0<=random.random() < 1.0
print(random.randint(1,10))    # random 정수형 숫자 추출(10 포함됨)
print(random.uniform(10,20))   # uniform(최소값,최대값) : 지정한 최소, 최대숫자 난수 발생(최대값 포함됨)
print(random.randrange(10))    # randrange(start,stop,step) : 최소, 최대, 차이값
print(random.choice([1,2,3,4,5,6,7]))  #choice(sequence) : 리스트,튜플,문자열
print(random.choice("python"))  #choice(sequence) : sequence는 리스트,튜플,문자열 
list1=list(range(1,8))
print(random.shuffle(list1))   #shuffle(sequence) : sequence를 랜덤으로 섞는다.
print(list1)
print(random.sample(list1,k=2))   #sample(sequence, 개수) : 리스트, 튜플, 문자열

```

```python
from random import random, randint, uniform, randrange, choice, shuffle, sample
#random 모듈내 기능을 import 하면 아래에서 random.은 삭제 가능

print(random())         #random() : 0.0 ~ 1.0 사이의 실수 중에서 난수 (0은 포함하지만 1은 미포함)
# 0.0<=random.random() < 1.0
print(randint(1,10))    # random 정수형 숫자 추출(10 포함됨)
print(uniform(10,20))   # uniform(최소값,최대값) : 지정한 최소, 최대숫자 난수 발생(최대값 포함됨)
print(randrange(10))    # randrange(start,stop,step) : 최소, 최대, 차이값
print(choice([1,2,3,4,5,6,7]))  #choice(sequence) : 리스트,튜플,문자열
print(choice("python"))  #choice(sequence) : sequence는 리스트,튜플,문자열 
list1=list(range(1,8))
print(shuffle(list1))   #shuffle(sequence) : sequence를 섞는다.
print(list1)
print(sample(list1,k=2))   #sample(sequence, 개수) : 리스트, 튜플, 문자열
```

```python
# from os.path import isdir
# os : 환경변수, 디렉터리, 파일 등의 OS 자원을 제어할 수 있게 해주는 모듈

import os

output = os.listdir(".")   # listdir(폴더명) : 지정된 폴더의 모든 파일과 폴더를 리스트로 가져옴. "."은 현재 폴더

for path in output:
    if os.path.isdir(path):  #is 붙으면 True나 False로 결과 보여줌. path가 디렉토리가 맞는지 확인. 맞으면 "폴더" 아니면 "파일"
        print("폴더".path)
    
    else:
        print("파일".path)
```

```python
import csv   #csv : csv파일들을 관리하기 편하게 하기 위해 만들어진 모듈 이름

with open("./data/sample1.csv","r") as f:
    reader = csv.DictReader(f)
    # print(reader)
    next(reader)
    for c in reader:
        for k, v in c.items():
            print(k,v)  
        # DictReader로 읽으면 딕셔너리 구조로 읽어옴
```

```python
import json   #json모듈 : json파일들을 관리하기 편하게 하기 위해 만들어진 모듈 이름
              # json파일 : 딕셔너리 구조와 비슷하게 생긴 형태의 텍스트 파일 (JavaScript에서 파생된 언어)

with open("./data/myinfo.json","r",encoding="utf-8") as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
# loads() : json형태(딕셔너리형태)로 작성된 문자열을 파이썬 자료형(딕셔너리)로 변환


str1 = """
{
    "name": "홍길동",
    "birth": "0727",
    "age": 20
}
"""  
#딕셔너리 구조로된 문자열을 딕셔너리로
data=json.loads(str1)
print(data)
print(type(data))

print(data['name'])
print(data.get('name'))
```

```python
# dumps() : 파이썬 자료형(딕셔너리)을 json구조(딕셔너리구조)의 문자열로 변환

dict1 = {
    "name": "홍길동",
    "birth": "0727",
    "age": 20
}

data=json.dumps(dict1)
print(data)
print(type(data))

```

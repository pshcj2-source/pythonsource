#### 클래스 (붕어빵 틀 설계도 같은 것, 객체에 팥, 피자, 슈크림을 담을 수 있음)

# data가 많아지면 위와 같이 기존 방법으로 관리가 어려움
# 설계도
# 클래스 이름은 대문자로 시작
# 메소드(==함수) : 클래스 내부에 작성되는 함수. 단독으로 불러다 쓸 수 없다.

```python
# 클래스는 일반변수, 리스트, 딕셔너리도 한번에 넣어서 관리 가능
# 클래스를 쓰기 전에 설계를 먼저 해야함
class Calculator:  # Class 이름 지정 
    def __init__(self):  #언더바 2개씩 되어있는 것 사이에 오는 것은 특수 기능을 가진 method, self라는 것은 무조건 가지고 있어야함 (클래스 안에 사용되는 함수를 method 라고 함)
        self.result=0

    def add(self,num):
        self.result += num
        return self.result

    def sub(self,num):
        self.result -= num
        return self.result
```

```python
# 계산기 2개 만들기

# 객체 생성  -> __init()__ 자동으로 호출됨
# 변수 = 클래스명()

calc1 = Calculator()
print(calc1.add(3))  #변수를 부르는 방법
print(calc1.add(4))

calc2 = Calculator()
print(calc2.add(13))
print(calc2.add(14))

```

#### 인스턴스(멤버) 변수
- 객체마다 따로 따로 가지는 데이터
- 'self.변수명" 형태로 작성'

```python
# 자동차 클래스 설계
# 속성 : 색상, 최대속도 => 멤버 변수로 만들기
# 동작 : 전진, 후진, 가속, 감속... => 멤버 메소드로 만들기

class Car:
    def __init__(self,color,speed):
        # 멤버 변수들은 init안 에다가 self.변수명 형태로 작성
        self.color=color
        self.speed=speed
    
    # 인스턴스(=멤버) 메소드(함수)
    def upSpeed(self, value) :
        self.speed += value

    def downSpeed(self, value):
        self.speed -= value

```

```python
# 클래스가 잘 만들어졌는지 확인은 객체를 생성해서 검증
# Car 객체 생성 방법
# 변수명 = 클래스명()

#  Car.__init__() 자동호출
car1 = Car("Red",50)
car2 = Car("Black",70)

print(car1)
print(car2)


# 결과값은 메모리 어디에 저장되어있는지 알려주는 결과
# <__main__.Car object at 0x00000120B4926BA0>
# <__main__.Car object at 0x00000120B47A5590>
```

```python
# 자동차 클래스 설계
# 속성 : 색상, 최대속도 => 멤버(=인스턴스) 변수로 만들기
# 동작 : 전진, 후진, 가속, 감속... => 멤버 메소드로 만들기

class Car:
    def __init__(self,color,speed):  # init은 인스턴스 변수 초기화 기능이 있음
        # 멤버 변수들은 init안 에다가 self랑 변수명 형태로 작성
        self.color=color
        self.speed=speed
    
    # 인스턴스(=멤버) 메소드(함수)
    def upSpeed(self, value) :
        self.speed += value

    def downSpeed(self, value):
        self.speed -= value

# str : "0x00000120B4926BA0" 형태를 보고싶은 형태로 바꿔주는 것
    def __str__(self):
        return f"Car(color={self.color}, speed={self.speed})"
```

```python
car1 = Car("Red",50)
car2 = Car("Black",70)

print(car1)
print(car2)

# str 함수가 정의되어서 원하는 형태로 출력됨.
```

#### 클래스 메소드
- 클래스 자체를 다루는 메소드(함수)
- 첫 번째 매개변수로 'self' 대신 'cls'를 받음
- 클래스 변수 관리

```python
class SelfTest:
    @classmethod
    def function1(cls):
        print("function1 called")

    def function2(self):
        print("function2 called")

test1 = SelfTest()
# test1.function1()
# 클래스이름.클래스메소드 형태로 접근
SelfTest.function1()
test1.function2()
```

```python
class Dog:

    # 클래스 변수 (공통적으로 사용할 변수)
    species = "강아지" #공통으로 적용되는 특징은 class변수로 지정
    count = 0

    def __init__(self,name,age):
        self.name=name
        self.age = age
        Dog.count += 1
        

    def bark(self):
        return f'{self.name} : 멍멍'

    @classmethod  #클래스 메소드는 @ 붙여줌
    def get_count(cls):
        return f"지금까지 만들어진 {cls.species}는 {cls.count}입니다."

# 변수 = 클래스명(이름, 나이)

dog1=Dog("멍뭉이",5)
dog2=Dog("바둑이",7)

print(dog1.name, dog1.age)
print(dog2.name, dog2.age)

print(dog1.bark())
print(dog2.bark())

print(Dog.get_count())
```

| 구 분 | 첫 인자 | 접근 가능 | 주 용도 |
|----|----|----|----|
| '인스턴스 메서드' | 'self 사용' | 인스턴스 변수, 클래스 변수 모두 | 객체별 동작 |
| '@classmethod' | 'cls 사용' | 클래스 변수만 | 클래스 단위 동작 |

```python
class Car:

    # 클래스 변수 (self 사용 하지 않음)
    car_count=0


    def __init__(self,speed,color="Black"):   #색상 정보가 안들어오는 경우 Black으로 기본값 설정. 기본값이 정해진 변수는 가장 뒤로
        self.color=color
        self.speed=speed
        Car.car_count += 1
        #셀프 안붙이면 공통으로 저장되는 공간으로 저장
    
    # 인스턴스(=멤버) 메소드(함수)
    def upSpeed(self, value) :
        self.speed += value

    def downSpeed(self, value):
        self.speed -= value

# str : "0x00000120B4926BA0" 형태를 보고싶은 형태로 바꿔주는 것
    def __str__(self):
        return f"Car(color={self.color}, speed={self.speed})"

    @staticmethod
    def hello():
        print("안녕하세요")

car1 = Car(80,"Red")
car2 = Car(70,"Withe")
car3 = Car(85)

Car.hello()
print(f"생산된 자동차 숫자는 {Car.car_count}대")
```

```python
# Student 클래스 작성
# 멤버변수 : 이름(name), 나이(age), 점수(score)
# 멤버 메서드 : study(hour) : 공부한 시간 * 2 의 결과를 점수에 추가
#             showInfo() : 이름:홍길동, 나이:20, 점수:86


class Student:
    def __init__(self,name,age,score):
        self.name = name
        self.age = age
        self.score = score

    def study(self, hour):
        self.score += 2 * hour

    def showInfo(self):
        return f'이름 : {self.name}, 나이 : {self.age}, 점수 : {self.score}'
            
stu1=Student("홍길동",20,80)
stu1.study(3)
stu1.showInfo()

```

```python
# Account 클래스 작성
# 멤버변수 : 계좌주(owner), 잔액(balance)
# 멤버 메서드 : deposit(money) : 잔액 = 잔액 + 입금액
#            : withdraw(money) : 잔액 = 잔액 - 출금액
#            : showBalance() : 현재 잔액 출력

class Account:
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance

    def deposit(self,money):
        self.balance = self.balance + money

    def withdraw(self,money):
        if self.balance < money:
            print("잔액부족")
        else:
            self.balance = self.balance - money

    def showBalance(self):
        print(f'현재잔액 : {self.balance}원')

account = Account("홍길동",10000)
account.deposit(20000)
account.withdraw(15000)
account.showBalance()

```

```python
# Book 클래스 작성
# 클래스 변수 : book_count
# 멤버변수 : 제목(title), 저자(author)
#   메서드 : showInfo() 출력 => 제목 : 파이썬
#                             저자 : 홍길동
# 현재 가지고 있는 책 권수 출력

class Book:
    book_count=0

    def __init__(self, title, author):
        self.title=title
        self.author=author
        Book.book_count += 1

    def showInfo(self):
        print(f'제목 : {self.title}\n저자 : {self.author}')
        
book1=Book("파이썬","홍길동")
book2=Book("자바","임꺽정")
book1.showInfo()
book2.showInfo()
print(f"발행된 책 {Book.book_count}권")
```

```python
# Order 클래스 작성
# 클래스 변수 : total_order
# 멤버 변수 : 메뉴(menu), 가격(price), 수량(count)
# 메소드 : getTotalPrice() = 가격 x 수량 : 총 주문금액 반환
#         showOrder() : 출력문 작성 (아래 형태)
#         메뉴 : 아메리카도
#         가격 : 4000원
#         수량 : 2개
#   총 주문금액 : 8000원

class Order:

    total_order=0

    def __init__(self, menu, price, count):
        self.menu=menu
        self.price=price
        self.count=count
        Order.total_order += 1

    def getTotalPrice(self):
        return self.price * self.count

    def showOrder(self):
        print(f"메뉴 : {self.menu}\n가격 : {self.price}\n수량 : {self.count}\n총 주문금액 : {self.getTotalPrice()}")


order1 = Order("아메리칸",4500, 2)
order2 = Order("카페라떼",4800, 1)

order1.showOrder()
print()
order2.showOrder()

print(f"전체 주문 수 : {Order.total_order}")
```

#### 상속
- 기존 클래스(부모/상위)의 멤버변수와 메서드를 새로운 클래스(자식/하위)가 그대로 물려받는 것
- 코드 재사용성, 확장

```python
class Parent:   # 기존 작성된 class 를 Parent라고 하면
    def __init__(self):
        self.value = "테스트"
        print("Parent 클래스의 __init__() 호출")

    def test(self):
        print("Parent 클래스의 test() 호출")

class Child(Parent): # 기존 작성된 Parent class를 사용하려면 괄호에 기존 class명을 넣어줌
    pass

child = Child()
child.test()
print(child.value)
```

```python

# supper() : 상위 클래스
class Child(Parent): 
    def __init__(self):
        super().__init__(self)  #super() : 자식 클래스에서 부모 클래스의 메서드나 속성에 접근하게 해주는 내장 함수
        print("Child 클래스의 __init__ 호출")


child = Child()
child.test()
print(child.value)
```

```python
class Animal:
    def __init__(self,name):
        self.name=name

    def eat(self):
        print(f"{self.name}가 먹이를 먹는다")

    def sound(self):
        print(f"{self.name}가 소리를 낸다")


# parent가 가지고 있는 것 외 일부 추가해서 class를 만들고 싶을떄
class Dog(Animal):
    def __init__(self, name, kind):
        super().__init__(name)
        self.kind = kind

    def fetch(self):
        print(f"{self.name}가 공을 물어옵니다.")

dog = Dog("초코","푸들")
dog.eat()
dog.sound()
dog.fetch()
```

```python
# Audio : 멤버변수 : power, volume / 메소드 : 전원 조절, 볼륨조절 + tune()
# TV 멤버변수 : power, volume, size / 메소드 : 전원 조절, 볼륨조절 + watch()
# 공통적인 것을 뽑아서 parent로 만들고 나머지 추가되는 것들을 개별 class에 넣는다.

class AudioVisual:
    def __init__(self,power,volume):
        self.power=power
        self.volume=volume

    def switch(self,on_off):
        self.power=on_off
    
    def setVolume(self,volume):
        self.volume=volume

class Audio(AudioVisual):
    def __init__(self, power, volume):
        super().__init__(power, volume)

    def tune(self):
        print("La la la..." if self.power else "turn it on")

class TV(AudioVisual):
    def __init__(self, power, volume,size):
            super().__init__(power, volume)
            self.size=size

    def watch(self):
        print("have fun!!" if self.power else "switch on")


tv = TV(False, 12, 50)
# tv 켜기
tv.switch(True)
# tv 보기
tv.watch()

audio = Audio(True,15)
audio.setVolume(10)
audio.tune()
audio.switch(False)
```

```python
# Employee 부모 클래스 : 
# 멤버 변수 : 이름(name), 급여(pay)
#     메소드 : showInfo() => 이름, 급여 출력

# Manager 자식 클래스
# 멤버 변수 : 보너스(bonus)
# 메소드 : getTotalPay() : 급여 + 보너스 출력

class Employee:
    def __init__(self,name,pay):
        self.name=name
        self.pay=pay

    def showInfo(self):
        print(f' 이름 : {self.name}, 급여 : {self.pay}')

class Manager(Employee):
    def __init__(self, name, pay, bonus):
        super().__init__(name, pay)
        self.bonus=bonus

    def getTotalPay(self):
        print(f"총 급여 : {self.pay+self.bonus}만원")

emp=Employee("홍길동",750)
emp.showInfo()

manager=Manager("성춘향",640,800)
manager.showInfo()
manager.getTotalPay()
```

```python
# Shape 부모 클래스 
# 메서드 : area() => "도형의 넓이" 출력

# Rectangle 자식 클래스
# 멤버 변수 : width, height
# 메서드 : area() => 직사각형의 넓이 출력

# Circle 자식 클래스
# 멤버 변수 : 반지름(radius)
#    메서드 : area() => 원의 넓이 출력

class Shape:
    def area(self):
        print("도형의 넓이")

class Rectangle(Shape):
    def __init__(self,width,height):
        super().__init__()
        self.width=width
        self.height=height
    
    def area(self):
        print(f'사각형 넓이: {self.width * self.height}')

class Circle(Shape):
    def __init__(self,radius):
        super().__init__()
        self.radius=radius
            
    def area(self):
        print(f'원의 넓이 : {self.radius * self.radius * 3.14}')

r = Rectangle(10,20)
r.area()

c = Circle(5)
c.area()
```

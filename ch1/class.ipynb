```python
#결과저장 변수  (계산기에서 여러가지 숫자 더하기 빼기 진행할떄 중간값이 화면에 보여지면서 더해지는거)
#계산기가 2대 일때

result1=0
result2=0

def add(num):
    global result1
    result1 += num   #result1 = result1 + num
    return result1

def add2(num):
    global result2
    result2 += num   
    return result2


print(add(3))
print(add(4))

print(add2(13))
print(add2(14))
```

#### 클래스 (붕어빵 틀 설계도 같은 것, 객체에 팥, 피자, 슈크림을 담을 수 있음)

# data가 많아지면 위와 같이 기존 방법으로 관리가 어려움
# 설계도
# 클래스 이름은 대문자로 시작
# 메소드(==함수) : 클래스 내부에 작성되는 함수. 단독으로 불러다 쓸 수 없다.

```python
class Calculator:
    def __init__(self):  #언더바 2개씩
        self.result=0

    def add(self,num):
        self.result += num
        return self.result

    def sub(self,num):
        self.result -= num
        return self.result
```

```python
# 계산기 2개 만들기

# 객체 생성  -> __init()__ 자동으로 호출됨
# 변수 = 클래스명()

calc1 = Calculator()
print(calc1.add(3))
print(calc1.add(4))

calc2 = Calculator()
print(calc2.add(13))
print(calc2.add(14))

```

```python
class Student:
    def __init__(self,name,number,grade,detail):
        self.name = name
        self.number = number
        self.grade = grade
        self.detail = detail

    # 인스턴스를 출력할 떄 출력 시 원하는 값들을 반환
    def __str__(self):
        return f"이름 : {self,name}, 반 : {self.number}, 학번 : {self.grade}, 상세정보 : {self.detail}"
```

```python
# 객체 생성 -> __init()__ 자동으로 호출됨
s1 = Student("kim",1,1,{'gender':'male','score1':95,'score2':88})
s2 = Student("Park",2,2,{'gender':'female','score1':75,'score2':68})
s3 = Student("Hong",2,1,{'gender':'male','score1':99,'score2':86})


# s1은 Student 클래시의 인스턴스 임
print(s1) # <__main__.Student object at 0x
print(s2) # <__main__.Stud
print(s3)
```

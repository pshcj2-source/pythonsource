```python
num=input("숫자입력")   #(상단에 뜨는 칸에 아무 숫자 입력)
type(num)    #숫자를 입력해도 결과값은 문자형으로 나옴
#사용자에게 숫자를 입력받은 후 +10 한 결과 출력
# 숫자가 아닌 문자를 잘못입력하는 경우 type 함수를 통해 확인하여 받아야함
# try ~ except 라는 기능 사용
if num.isdigit():
    num=int(num)+10
    print(num)
else:
    print("숫자를 입력하세요")
```

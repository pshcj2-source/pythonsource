```python
# try ~ except   # 예외가 발생할 수 있는 가능성이 있는 구문을 try~except 구문 안에 넣어주기


names = ["Kim","Park","Lee"]


try:
    z= "Let"  # 예외가 발생할 수 있는 가능성이 있는 구문
    x = names.index(z)
    print(f"{z} Found it!:")
except:
    print("에러 발생")
```

```python
# 사용자로부터 숫자(반지름)을 입력받아 원 둘레, 넓이를 출력하는 프로그램 만들기
# 숫자가 아닌 문자가 입력될 경우 오류가 뜸

try:
    number=int(input("반지름 입력"))
    print(f'둘레 : {2*3.14*number}')
    print(f'넓이 : {3.14*number**2}')

except 
    print("숫자를 입력해 주세요")


```

```python
# except + 에러명
# 에러나는 구문 2가지 조합. 에러의 종류에 따라 출력 메세지 다르게 처리 가능
try:  
    a=[1,2]   
    print(a[3])   # IndexError
    4/0   # ZeroDivisionError
except ZeroDivisionError as e:
    print(e)
except IndexError:
    print("인덱스 확인")
```

```python
# 오류 강제 발생 (다음 작업을 못하게 하기위해서)

class Bird:
    def fly(self):
        raise NotImplementedError  # raise : 발생시키고 싶은 오류를 발생시킴 (강제로 무엇인가를 강요할 때)

class Eagle(Bird):
    def fly(self):   # fly를 재정의하게 하기 위해서
        print("very fast")

eagle=Eagle()
eagle.fly()
```

```python
# 예외 클래스 생성 가능
class MyError(Exception):
    def __str_(self):
        return "허용되지 않는 별병입니다."

def say_nick(nick):
    if nick == "바보":
        raise MyError()
    print(nick)


say_nick("천사")
say_nick("바보")
```

```python
# 예외 클래스 생성 가능
class MyError(Exception):
    def __str_(self):
        return "허용되지 않는 별병입니다."

def say_nick(nick):
    if nick == "바보":
        raise MyError()
    print(nick)

try:
    say_nick("천사")
    say_nick("바보")

except MyError as e:
    print(e)
```

```python
# 문제 1. 랜덤 비밀번호 생성기

# 사용자로부터 생성할 비밀번호의 길이를 입력받아 랜덤 비밀번호를 생성하는 프로그램 작성
# Choice()
# 비밀번호 길이는 4자 이상만 허용
# 숫자가 아닌 값을 입력하면 예외 처리한다.

Chars = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789!@#$%"

import random
while True:
    try:
        pwd_num=int(input("비밀번호 길이 : "))

        if pwd_num<4:
            print("비밀번호는 최소 4자리 이상 입니다.")
            continue
        break

    except ValueError:
        print("숫자만 입력하세요")

# 리스트 컴프리헨션 활용방법
# password = []
# for i in range(pwd_num):
# password.append(random.choice(Chars))

password=[random.choice(Chars) for _ in range(pwd_num)]  #i 쓰일게 없으면 대신 _ 로 사용

# print(f'랜덤 비밀번호 : )
# 리스트 -> 문자열로 변경 (join)
print(f"랜덤 비밀번호 : {"".join(password)}")



```

```python
# 나이 계산 프로그램
# 생년월일을 입력받아 현재 나이를 계산하는 프로그램 작성
# 입력형식 2001-08-25

import datetime
# 오늘날짜
now = datetime.datetime.now()
print(type(now))

try:
    # 생년월일 입력
    birth_str=input("생일을 입력하세요(YYYY-MM-DD)")
    birth=datetime.datetime.strptime(birth_str,"%Y-%m-%d")
    print(type(birth))

    # 년도만 추출
    age=now.year - birth.year
    if (now.month,now.day)<(birth.month,birth.day):
        age-+1
    print("나이", age)
except:
    print("날짜 형식이 올바르지 않습니다.")

```

```python
# 다음과 같은 csv 파일이 있다.
# 이름,국어,영어,수학
# 홍길동,91,80,90
# 김철수,92,89,77
# 이영희,90,88,78
# 성춘향,95,80,89

import csv

with open("./data/student.csv","w",encoding="utf-8",newline="") as f:
    headers = ['이름','국어','영어','수학']
    writer = csv.DictWriter(f, fieldnames=headers)
    # 헤더 작성
    writer.writeheader()
    writer.writerow({"이름":"홍길동","국어":91,"영어":80,"수학":90})
    writer.writerow({"이름":"김철수","국어":92,"영어":89,"수학":77})
    writer.writerow({"이름":"이영희","국어":90,"영어":88,"수학":78})
    writer.writerow({"이름":"성춘향","국어":95,"영어":80,"수학":89})

# csv 파일을 읽어 학생별 평균 점수 출력
# 홍길동 : 80.0

# csv.DictReader() 사용, 평균은 둘째자리까지 표시


with open("./data/student.csv","r",encoding="utf-8",newline="") as f:
    reader = csv.DictReader(f)
    # print(reader)

    for row in reader:
        print(row) # {'이름': '홍길동', '국어':'88', '영어':'92','수학':'70'}
        avg = (int(row.get('국어')) + int(row.get('영어'))+int(row.get('수학')))/3
        print(f"{row.get('이름')} : {avg:.2f}")
```

```python
# 주차장 프로그램

# 사용자에게 메뉴 제공
# [1] 주차 | [2] 출차 | [3] 종료
# 자동차 A ~ E (총 5대 주차가능)
# 사용자는 메뉴 선택

# 1 (버튼입력 1번)
# 화면 출력 : A 자동차 입차, 주차장 상태 --> ['A']
# 1 (버튼입력 1번)
# 화면 출력 : B 자동차 입차, 주차장 상태 --> ['A','B']
# 2 (버튼입력 2번)
# 화면 출력 : B 자동차 출차, 주차장 상태 --> ['A']  (나중에 들어온 차부터 빠짐)
# ~~~
# E 자동차입차, 주차장 상태 --> ['A','B','C','D','E']  #만차 상태
# 1 (버튼입력 1번)
# 화면 출력 : 주차장 만차

parking_lot=[]   #리스트형
top, car_name=0,'A'

  
while True:
    no=int(input("[1] 주차 | [2] 출차 | [3] 종료 : "))    
    
    if no <=3:
        if no ==1:
            # top>=5 이상 시 주차장 만차(아무것도 안함)
            if top>=5:
                print("주차장 만차")
            else:
                parking_lot.append(car_name)
                 # 출력 : A 자동차 입차. 주차장 상태 -->['A']
                print(f'{car_name} 자동차 입차, 주차장 상태 --> {parking_lot}')
                car_name = chr(ord(car_name)+1)   #chr : 숫자에 대응되는 문자 호출 #ord : 문자에 대응되는 숫자 호출
                top=top+1
            # parking_lot.append(car_name)
            # car_name=chr(ord(car_name)+1)
            # top=top+1
            pass          #함수가 준비 중일 때(완성이 안되었을때) 실행시켜도 에러가 뜨지않게
        elif no==2:
            # top>0일때 리스트 뒤부터 제거 : pop()            
            if top>0:
                outCar=parking_lot.pop()
                print(f'{outCar} 자동차 출차. 주차장 상태 --> {parking_lot}')
                top = top-1
                car_name=chr(ord(car_name)-1)
            # B 자동차 출차, 주차장 상태 -->['A']
            # top-1, car_name-1
            # top<1이면 "빠져나갈 자동차 없음"
            else:
                print("빠져나갈 자동차가 없음")
        else:
            print("프로그램 종료")
            break
    else:
        print("번호를 확인해 주세요")    

    
    pass
```

```python
# 1~6이라는 숫자를 가지는 리스트를 생성

# numbers = [1,2,3,4,5,6]
# 빈 리스트 생성 후 1~6 숫자를 가지는 리스트
numbers = []
numbers.append(1)
numbers.append(2)
numbers.append(3)
numbers.append(4)
numbers.append(5)
numbers.append(6)

#-> 반복문으로 간단하게 작성
numbers = []

for i in range(1,7):
    numbers.append(i)

print(numbers)

numbers2 = list(range(1,7))
print(numbers2)
```

```python
# 리스트 컴프리핸션 방법을 활용한 새로운 리스트 생성

print([i for i in range(1,7)])
print([i*i for i in range(5)])
print([i*2 for i in [1,3,7,9,16]])
print([[i, i+1, i+2] for i in [1,2,3]])

# 1~10 사이의 숫자 중에서 홀수 리스트 생성
odd_list=[]
for i in range(1,11):
    if i%2==1:
        odd_list.append(i)
odd_list


# 더 간략하게 하는 방법
odd_list2 = [i for i in range(1, 11) if i % 2 == 1]
print(odd_list2)
```

```python
q_list=["값","을","병","정","경"]

# q_list에서 정을 제외한 요소를 갖는 새로운 리스트 생성
new_list=[]
for i in q_list:
    if i%2~="정":
        new_list.append(i)
print(new_list)


# 리스트 컴프리헨션 사용
q_list2=[i for i in q_list if i!="정"]  #range는 숫자에 적용되는 함수

print(q_list2)
```

```python
# 5글자 이상의 단어만 새로운 리스트로 생성
q1=['nice','study','python','anaconda','!']
[x for x in q1 if len(x)>=5]
```

```python
# 소문자만 추출 후 새로운 리스트로 생성
q2=['A','B','C','D','e','F','G','f']

# is~~함수 : ~~가 맞는지 물어보는 함수
# islower() : 소문자인지 묻는것, isupper() : 대문자인지 묻는것,  isalum() : 숫자인지 영어인지 묻는것

q3=[i for i in q2 if i.islower()]
print(q3)
```

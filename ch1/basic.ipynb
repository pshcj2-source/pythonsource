### 문자열 포매팅
- 문자열 안에 어떤 값을 삽입하는 방법
|코드|설명|
|---|---|
|`%s`|문자열(string)|
|`%d`|정수문자열(integer)|
|`%c`|문자열(character)|
|`%f`|부동 소수(Floating-point)|
|`%o`|8진수|
|`%x`|16진수|
|`%%`|%문자 자체|



```python
a="I eat %d apples" %3
print(a)
```

```python
a="I eat %s apples" %"three"
print(a)
# %S : 어떤 값이든 대입 가능
```

```python
# 공백제거
print('hi'=='  hi  ')   # ==는 둘이 같은가?
a='  hi  '
print(a.lstrip())  #왼쪽 공백 제거
print(a.rstrip())  #오른쪽 공백 제거
print(a.strip())  #양쪽 공백 제거
print('hi'==a.strip())
print(a)
```

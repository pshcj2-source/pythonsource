```python
s1=3
print(type(s1))

# 다른 타입으로 변경 ()
# int, float => str (숫자->문자) str함수
print(type(str(s1)))
# 문자 -> 숫자
print(type(int(str(s1))))
print(type(float(str(s1))))

s2=True
# bool -> str (불->문자) str함수
print(type(s2))
print(type(str(s2)))

# 문자->숫자형
s3='1'
print(int(s3)) #정수형
print(float(s3)) #실수형
print(bool(s3)) # bool은 1이면 True, 0이면 False

s4=0
print(bool(s4))
s5="0"
print(float(s5))
print(bool(float(s5)))
```

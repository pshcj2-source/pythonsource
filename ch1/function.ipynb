```python
def add_mul(choice, *args):  #Choice, a : 일반변수 *args : 가변변수
    pass
# 일반 매개변수와 *args를 혼용해서 사용하기 위해서는 일반 매개변수를 앞쪽으로 선언해야 함

# + 더하기, * 곱하기 실행
def add_mul(choice, *args):  # 더하기 곱하기는 choice로 선택
    if choice == "add":
        result=0
        for i in args:
            result=result+i
    
    elif choice == "mul":
        result=1
        for i in args:
            result=result*i
    
    return result

```

```python
# 키워드 매개변수 (key=value)
# **변수명  (*가 두개면 Key=value 형태의 가변변수를 받는다. -> 딕셔너리 형태)

# 딕셔너리 == 키워드 매개변수
def print_kwargs(**kwargs):
    print(kwargs)

    # key값만 출력
    for k in kwargs.keys():
        print(k)
    
    # value값만 출력
    for v in kwargs.value():
        print(v)

    # key, value값 같이 출력

    for k,v in kwargs.items():
        print(k,v)
```

```python
print_kwargs(name="kim",age="15")
print_kwargs(name="park",age="25", addr="서울")
print_kwargs(name="hong",age=30, addr="경기", tel="010-1234-5678")
```

```python
# 사칙연산 함수
# 결과 리턴
def calculator(num1, num2, op):
    # if-elif
    if op=="+":
        return num1 + num2
    elif op=="-":
        return num1 - num2
    elif op=="*":
        return num1 * num2
    elif op=="/":
        return num1 / num2
    elif op=="//":
        return num1 // num2
    

print(calculator(10,20,"+"))
print(calculator(10,20,"-"))
print(calculator(10,20,"*"))
print(calculator(10,20,"/"))
print(calculator(10,20,"//"))
```

```python
strings=["bob","charles","alexander3","teddy"]
strings.sort() # sort함수의 기본정렬은 첫 글자의 알파벳 순서, Key를 설정하면 정렬기준을 바꿀 수 있음.

# 정렬 : 문자 길이에 따라

def str_len(s):
    return len(s)

# 정렬 : 문자 길이 (len())에 따라
# strings.sort() # 기본정렬 알파벳 순서

strings.sort(key=str_len, reverse=True)

# lambda 함수로 바꾸기
strings.sort(key=lambda, s:len(s), reverse=True)
```

```python
# 1000 명의 이름, 키, 몸무게를 저장해 둔 info.txt 를 읽은 후 
# BMI 지수 계산 (BMI = 체중/키^2)
# bmi가 18.5 미만일 때 저체중
# bmi가 18.5~22.9 정상체중
# bmi가 23~24.9 과체중
# bmi가 25 이상 비만

# 표시방법
# 이름 : OO
# 몸무게 : xx
# 키 : xxx
# BMI : xx.x
# 결과 : 정상체중


with open("./data/info.txt","r",encoding="utf-8") as f:
    for line in f:
        name,weight,height=line.strip().split(", ")   #strip은 마지막 줄 엔터때문에
        weight=int(weight)
        height=int(height)

        #print(name,weight,height)
    
        bmi=weight/(height/100)**2
        
    

        if bmi<18.5:
            result = "저체중"
        elif bmi<=22.9:
            result = "정상체중"
        elif bmi<=25:
            result = "과체중"
        else:
            result = "비만"
        

        print("\n".join(["이름 : {}", "몸무게 : {}","키 : {}","BMI :{:.1f}","결과 : {}"]).format(name,weight,height,bmi,result))
        print()  #사람간 한칸씩 간격 조정

```

```python
# origin.txt 파일을 읽어서 암호화 시킨 후 encr.txt로 저장  
# origin.txt 작성 후 저장을 해줘야 함
# 1. 암호화 선택 : origin.txt 읽어서 변수에 담기 -> 그걸 chr(ord("안")+100) 저장


while True:
    no=int(input("1.암호화 | 2.복호화 | 3.종료 중 원하는 메뉴의 번호를 입력하세요"))

    if no==1:
        with open("./data/origin.txt","r",encoding="utf-8") as f:
            contents=f.read()
           
        with open("./data/encr.txt","w",encoding="utf-8") as f:
            for c in contents:
                f.write(chr(ord(c)+100))
    elif no==2:
        with open("./data/encr.txt","r",encoding="utf-8") as f:
            contents=f.read()
            for c in range(len(contents)):
                print(chr(ord(contents[c])-100), end="")    #컨텐츠 길이만큼 index번호 돌아가면서 빼기
    else:
        break
```
